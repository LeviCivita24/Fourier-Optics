```python
import skimage as sk
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import scipy.ndimage as sc
from PIL import Image as im
a=im.open('./3.bmp')
print(np.max(a))
a=np.double(a)/255


plt.figure()
plt.imshow(a,cmap=plt.cm.gray)
b1=a*0.8
b1[1,1]=1

b1=np.uint8(b1*255)
plt.figure()
plt.imshow(b1,cmap=plt.cm.gray)
b2=sc.gaussian_filter(a,3)
b2=np.uint8(b2*255)
plt.figure()
plt.imshow(b2,cmap=plt.cm.gray)
[n,m]=np.shape(a)
b3=(a*0.8+0.2*np.random.rand(n,m))*1
b3=np.uint8(b3*255)
plt.figure()
plt.imshow(b3,cmap=plt.cm.gray)
a=np.uint8(a*255)
```

```python
import skimage.metrics
c1=sk.metrics.mean_squared_error(a,b1)
c2=sk.metrics.mean_squared_error(a,b2)
c3=sk.metrics.mean_squared_error(a,b3)

print('El mse para la imagen con menor intensidad es :', c1)
print('El mse para la imagen con filtro gaussiano es :', c2)
print('El mse para la imagen con ruido aleatorio es:', c3)
```

```python
c1=sk.metrics.peak_signal_noise_ratio(a,b1)
c2=sk.metrics.peak_signal_noise_ratio(a,b2)
c3=sk.metrics.peak_signal_noise_ratio(a,b3)

print('El psnr para la imagen con menor intensidad es :', c1)
print('El psnr para la imagen con filtro gaussiano es :', c2)
print('El psnr para la imagen con ruido aleatorio es:', c3)
```

```python
def corr2(A,B):
    e=0
    s1=0
    s2=0
    Am=np.mean(A)
    Bm=np.mean(B)
    [n,m]=np.shape(A)
    for i in range(0,n):
        for j in range(0,m):
            e=e+(A[i,j]-Am)*(B[i,j]-Bm)
            s1=s1+(A[i,j]-Am)**2
            s2=s2+(B[i,j]-Bm)**2
    s=np.sqrt(s1*s2)
    return e/s
c1=corr2(a,b1)
c2=corr2(a,b2)
c3=corr2(a,b3)

print('El cc para la imagen con menor intensidad es :', c1)
print('El cc para la imagen con filtro gaussiano es :', c2)
print('El cc para la imagen con ruido aleatorio es:', c3)
```

```python
c1=sk.metrics.structural_similarity(a,b1)
c2=sk.metrics.structural_similarity(a,b2)
c3=sk.metrics.structural_similarity(a,b3)

print('El ssim para la imagen con menor intensidad es :', c1)
print('El ssim para la imagen con filtro gaussiano es :', c2)
print('El ssim para la imagen con ruido aleatorio es:', c3)
```
